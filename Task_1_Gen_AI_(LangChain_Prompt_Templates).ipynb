```python
from langchain_core.prompts import PromptTemplate

# Define the template string with a placeholder
template_str = "Explain {topic} in simple terms for beginners"

# Create the LangChain PromptTemplate
prompt_template = PromptTemplate(input_variables=["topic"], template=template_str)

# Example usage
topic = "Quantum Computing"
formatted_prompt = prompt_template.format(topic=topic)
print(formatted_prompt)
```

```python
# Template: Explain {topic} for {audience} in a {tone} tone
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Case Example: Python | kids | fun
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
```

```python
teaching_prompt = PromptTemplate.from_template("Explain {topic} clearly step by step")
interview_prompt = PromptTemplate.from_template("Ask 3 questions about {topic}")
storytelling_prompt = PromptTemplate.from_template("Explain {topic} as a story")

topic_val = "Machine Learning"
print(teaching_prompt.format(topic=topic_val))
print(interview_prompt.format(topic=topic_val))
print(storytelling_prompt.format(topic=topic_val))
```

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful {role}."),
    ("user", "Explain {topic}")
])

# Example: role = 'teacher', topic = 'Neural Networks'
formatted_chat = chat_template.format_messages(role="teacher", topic="Neural Networks")
print(formatted_chat)
```

```python
def validate_inputs(audience, tone):
    allowed_audience = ["beginner", "intermediate", "expert"]
    allowed_tone = ["formal", "casual", "fun"]

    # Assign defaults if invalid
    final_audience = audience if audience in allowed_audience else "beginner"
    final_tone = tone if tone in allowed_tone else "formal"

    return final_audience, final_tone
```

```python
def generate_prompt(topic, audience, tone, style):
    # Validate inputs (from Task 5)
    valid_audience, valid_tone = validate_inputs(audience, tone)

    # Select style (Task 3 logic)
    styles = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style.",
        "interview": "Ask questions about {topic} for an {audience} in a {tone} interview style.",
        "storytelling": "Tell a story about {topic} for {audience} in a {tone} style."
    }

    template_str = styles.get(style, styles["teaching"])
    prompt_obj = PromptTemplate.from_template(template_str)

    return prompt_obj.format(topic=topic, audience=valid_audience, tone=valid_tone)

# Example Output
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
